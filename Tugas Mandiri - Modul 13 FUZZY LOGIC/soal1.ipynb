### Import library dan pendefinisian variabel simbolik

```python
import sympy as sp

x = sp.symbols("x")
```

Library SymPy, yaitu library Python yang sering dipakai untuk perhitungan matematika simbolik.

Variabel x dibuat sebagai variabel simbolik. Variabel ini nantinya digunakan sebagai dasar dalam membentuk fungsi keanggotaan fuzzy. Karena fungsi keanggotaan seperti suhu dingin, suhu sejuk, kelembaban lembab, dan kecepatan kipas dibuat dalam bentuk fungsi terhadap nilai x, maka x perlu didefinisikan terlebih dahulu.

### Fungsi trimf

```python
def trimf(a, b, c):
    return sp.Piecewise(
        (0, x <= a),
        ((x - a) / (b - a), x <= b),
        ((c - x) / (c - b), x < c),
        (0, True)
    )
```

Fungsi trimf, yaitu singkatan dari triangular membership function atau fungsi keanggotaan berbentuk segitiga.

Fungsi ini menerima tiga parameter, yaitu a, b, dan c. Parameter a adalah batas awal, b adalah titik puncak, dan c adalah batas akhir. Nilai keanggotaan akan bernilai 0 sebelum titik a, naik secara bertahap dari a ke b, mencapai nilai maksimum 1 di sekitar titik b, lalu turun kembali menuju 0 sampai titik c.

Fungsi ini digunakan untuk kategori fuzzy yang memiliki bentuk naik lalu turun, misalnya kategori “Sejuk”, “Normal”, “Pelan”, “Sedang”, dan “Cepat”.

### Fungsi trapmf

```python
def trapmf(a, b, c, d):
    if a == b:
        return sp.Piecewise(
            (0, x < a),
            (1, x <= c),
            ((d - x) / (d - c), x < d),
            (0, True)
        )
    if c == d:
        return sp.Piecewise(
            (0, x <= a),
            ((x - a) / (b - a), x < b),
            (1, x <= d),
            (0, True)
        )
    return sp.Piecewise(
        (0, x <= a),
        ((x - a) / (b - a), x < b),
        (1, x <= c),
        ((d - x) / (d - c), x < d),
        (0, True)
    )
```

Fungsi trapmf, yaitu trapezoidal membership function atau fungsi keanggotaan berbentuk trapesium.

Fungsi ini menerima empat parameter, yaitu a, b, c, dan d. Secara umum, fungsi ini memiliki empat bagian. Bagian pertama bernilai 0, bagian kedua naik secara bertahap, bagian ketiga bernilai penuh 1, dan bagian keempat turun kembali ke 0.

Namun, fungsi ini juga menangani dua kondisi khusus.

Kondisi pertama adalah ketika a == b. Artinya, fungsi langsung berada pada nilai maksimum sejak awal batas. Ini cocok untuk kategori seperti “Dingin”, “Kering”, atau “Mati”, karena kategori tersebut dimulai dari nilai paling rendah dan langsung memiliki derajat keanggotaan penuh.

Kondisi kedua adalah ketika c == d. Artinya, fungsi tetap bernilai maksimum hingga batas akhir. Ini cocok untuk kategori seperti “Panas”, “Lembab”, atau “Maksimal”, karena kategori tersebut berada di ujung kanan rentang nilai.

### Fungsi degree

```python
def degree(expr, nilai):
    return float(expr.subs(x, nilai))
```

Fungsi degree, yaitu fungsi untuk menghitung nilai derajat keanggotaan dari sebuah fungsi fuzzy.

Parameter expr berisi fungsi keanggotaan fuzzy, sedangkan nilai adalah nilai input yang ingin diuji. Program mengganti variabel simbolik x dengan nilai aktual, lalu hasilnya dikonversi menjadi angka desimal bertipe float.

Contohnya, jika input suhu adalah 28 derajat, maka fungsi ini dapat digunakan untuk mencari seberapa besar derajat keanggotaan suhu tersebut terhadap kategori “Sejuk” atau “Panas”.

### Fungsi Input Range

```python
def input_range(teks, minimum, maksimum):
    while True:
        try:
            nilai = float(input(teks))
            if minimum <= nilai <= maksimum:
                return nilai
            print(f"Input harus dalam range {minimum}-{maksimum}.")
        except ValueError:
            print("Input harus berupa angka.")
```

Fungsi input_range, yaitu fungsi untuk menerima input angka dari pengguna sekaligus memastikan nilainya berada dalam rentang yang diperbolehkan.

Fungsi ini berjalan dalam perulangan sampai pengguna memasukkan nilai yang valid. Jika input berupa angka dan berada di antara batas minimum serta maksimum, maka nilai tersebut dikembalikan. Jika nilai berada di luar rentang, program menampilkan pesan bahwa input harus sesuai range. Jika input bukan angka, program menampilkan pesan bahwa input harus berupa angka.

Fungsi ini digunakan agar program tidak memproses data yang tidak logis. Misalnya, suhu ruangan dibatasi antara 0 sampai 50 derajat, sedangkan kelembaban dibatasi antara 0 sampai 100 persen.

### Fungsi centroid, yaitu fungsi untuk melakukan proses defuzzifikasi.

```python
def centroid(output_mf, aturan, start, end, step):
    total_titik = int((end - start) / step) + 1
    pembilang = 0.0
    penyebut = 0.0

    for i in range(total_titik):
        nilai = start + i * step
        mu = max(min(w, degree(output_mf[label], nilai)) for label, w in aturan)
        pembilang += nilai * mu
        penyebut += mu

    return pembilang / penyebut if penyebut != 0 else 0.0
```

Dalam sistem fuzzy, hasil inferensi masih berupa derajat keanggotaan pada beberapa kategori output. Agar hasil akhirnya menjadi satu angka pasti, digunakan metode defuzzifikasi. Pada program ini, metode yang digunakan adalah metode centroid.

Fungsi ini menerima beberapa parameter. output_mf berisi fungsi keanggotaan output, aturan berisi hasil aturan fuzzy beserta kekuatan aktifnya, start dan end adalah batas rentang output, sedangkan step adalah jarak antar titik perhitungan.

Program menghitung banyak titik dari rentang output. Dalam kasus ini, rentangnya 0 sampai 100 dengan step 0.1, sehingga jumlah titiknya adalah 1001 titik.

Pada setiap titik, program menghitung nilai keanggotaan output berdasarkan aturan yang aktif. Nilai min(w, degree(...)) digunakan untuk membatasi fungsi output sesuai firing strength aturan. Kemudian max(...) digunakan untuk menggabungkan semua aturan fuzzy yang menghasilkan output.

Variabel pembilang menghitung jumlah nilai output dikalikan derajat keanggotaannya. Variabel penyebut menghitung total derajat keanggotaan. Hasil akhir centroid diperoleh dari pembilang dibagi penyebut.

Jika penyebut bernilai 0, artinya tidak ada aturan yang aktif, maka fungsi mengembalikan nilai 0.

```python
def interpretasi(nilai):
    if nilai <= 25:
        return "Mati"
    if nilai <= 45:
        return "Pelan"
    if nilai <= 65:
        return "Sedang"
    if nilai <= 85:
        return "Cepat"
    return "Maksimal"
```

Blok ini membuat fungsi interpretasi, yaitu fungsi untuk mengubah nilai numerik kecepatan kipas menjadi kategori bahasa.

Jika nilai hasil defuzzifikasi kurang dari atau sama dengan 25, maka interpretasinya “Mati”. Jika nilainya sampai 45, maka “Pelan”. Jika sampai 65, maka “Sedang”. Jika sampai 85, maka “Cepat”. Jika lebih dari 85, maka “Maksimal”.

### Input

```python
suhu_input = input_range("Masukkan suhu ruangan (0-50): ", 0, 50)
kelembaban_input = input_range("Masukkan kelembaban ruangan (0-100): ", 0, 100)
```

memasukkan dua nilai, yaitu suhu ruangan dan kelembaban ruangan.

Input suhu dibatasi dari 0 sampai 50, sedangkan input kelembaban dibatasi dari 0 sampai 100. Kedua input menggunakan fungsi input_range, sehingga program akan memastikan bahwa nilai yang dimasukkan valid.

### Fungsi keanggotaan untuk tiga variabel fuzzy, yaitu suhu, kelembaban, dan kecepatan kipas.

```python
suhu_mf = {
    "Dingin": trapmf(0, 0, 10, 20),
    "Sejuk": trimf(15, 22, 30),
    "Panas": trapmf(25, 35, 50, 50)
}

kelembaban_mf = {
    "Kering": trapmf(0, 0, 30, 50),
    "Normal": trimf(35, 50, 65),
    "Lembab": trapmf(55, 70, 100, 100)
}

kipas_mf = {
    "Mati": trapmf(0, 0, 10, 25),
    "Pelan": trimf(15, 35, 55),
    "Sedang": trimf(40, 55, 70),
    "Cepat": trimf(55, 75, 90),
    "Maksimal": trapmf(80, 90, 100, 100)
}

suhu = {k: degree(v, suhu_input) for k, v in suhu_mf.items()}
kelembaban = {k: degree(v, kelembaban_input) for k, v in kelembaban_mf.items()}
```

Pada variabel suhu, terdapat tiga kategori. “Dingin” menggunakan fungsi trapesium, “Sejuk” menggunakan fungsi segitiga, dan “Panas” menggunakan fungsi trapesium. Ini menggambarkan bahwa suhu rendah, sedang, dan tinggi memiliki bentuk keanggotaan yang berbeda.

Pada variabel kelembaban, terdapat tiga kategori. “Kering” menggunakan fungsi trapesium, “Normal” menggunakan fungsi segitiga, dan “Lembab” menggunakan fungsi trapesium.

Pada variabel output kecepatan kipas, terdapat lima kategori. Kategori tersebut adalah “Mati”, “Pelan”, “Sedang”, “Cepat”, dan “Maksimal”. Kategori ekstrem seperti “Mati” dan “Maksimal” menggunakan trapesium, sedangkan kategori tengah menggunakan segitiga.

Setelah fungsi keanggotaan didefinisikan, program menghitung derajat keanggotaan suhu input terhadap semua kategori suhu. Program juga menghitung derajat keanggotaan kelembaban input terhadap semua kategori kelembaban.

### Mendefinisikan sembilan aturan fuzzy.

```python
rules = [
    ("R1", "Dingin", "Kering", "Mati"),
    ("R2", "Dingin", "Normal", "Mati"),
    ("R3", "Dingin", "Lembab", "Pelan"),
    ("R4", "Sejuk", "Kering", "Pelan"),
    ("R5", "Sejuk", "Normal", "Sedang"),
    ("R6", "Sejuk", "Lembab", "Cepat"),
    ("R7", "Panas", "Kering", "Sedang"),
    ("R8", "Panas", "Normal", "Cepat"),
    ("R9", "Panas", "Lembab", "Maksimal")
]
```



Setiap aturan terdiri dari kode aturan, kategori suhu, kategori kelembaban, dan kategori output kecepatan kipas. Aturan ini menggunakan pola logika sederhana: jika suhu berada pada kategori tertentu dan kelembaban berada pada kategori tertentu, maka kipas berada pada kategori tertentu.

Contohnya, jika suhu “Dingin” dan kelembaban “Kering”, maka kipas “Mati”. Jika suhu “Panas” dan kelembaban “Lembab”, maka kipas “Maksimal”.

Karena suhu memiliki tiga kategori dan kelembaban juga memiliki tiga kategori, maka total kombinasi aturannya adalah 3 x 3, yaitu sembilan aturan.

### Perhitungan firing strength, centroid, dan aturan aktif

```python
firing = []
for kode, s, k, output in rules:
    w = min(suhu[s], kelembaban[k])
    firing.append((kode, s, k, output, w))

hasil = centroid(kipas_mf, [(output, w) for _, _, _, output, w in firing], 0, 100, 0.1)
aktif = [kode for kode, _, _, _, w in firing if w > 0]
```

Program membuat list kosong bernama firing. Kemudian setiap aturan diproses satu per satu. Untuk setiap aturan, program mengambil derajat keanggotaan suhu dan derajat keanggotaan kelembaban sesuai kategori pada aturan tersebut.

Nilai firing strength dihitung menggunakan operasi minimum. Hal ini sesuai dengan operator AND dalam logika fuzzy, yaitu mengambil nilai paling kecil dari dua derajat keanggotaan.

Setiap hasil aturan kemudian disimpan ke dalam list firing, lengkap dengan kode aturan, kategori suhu, kategori kelembaban, output, dan nilai firing strength.

Setelah itu, program menghitung hasil akhir menggunakan fungsi centroid. Output fuzzy yang sudah dipotong berdasarkan firing strength diproses menjadi satu nilai tegas dalam rentang 0 sampai 100.

Program juga membuat daftar aktif, yaitu daftar aturan yang memiliki firing strength lebih dari 0. Aturan dengan nilai lebih dari 0 dianggap aktif karena ikut memengaruhi hasil akhir.
