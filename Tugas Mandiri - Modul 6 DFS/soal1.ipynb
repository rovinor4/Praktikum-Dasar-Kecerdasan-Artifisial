## Penjelasan Dan Output Soal 1

```python
import matplotlib.pyplot as plt
import networkx as nx
```

Kode ini mengimpor library yang dibutuhkan. matplotlib.pyplot digunakan untuk menampilkan grafik, sedangkan networkx digunakan untuk membuat dan memproses graph.

```python
G = nx.Graph()

G.add_edges_from([
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("B", "E"),
    ("C", "F"),
    ("C", "G")
])
```

Kode ini membuat graph tidak berarah menggunakan nx.Graph(). Setelah itu, beberapa edge ditambahkan untuk menghubungkan node seperti A ke B, A ke C, B ke D, dan seterusnya.

```python
print("Urutan DFS tanpa batas:")
for edge in nx.dfs_edges(G, source="A"):
    print(edge)
```

Kode ini menampilkan urutan DFS tanpa batas kedalaman. Proses DFS dimulai dari node A, lalu menelusuri node-node yang terhubung secara mendalam terlebih dahulu.

```python
print("\nUrutan DFS dengan depth_limit=2:")
for edge in nx.dfs_edges(G, source="A", depth_limit=2):
    print(edge)
```

Kode ini menampilkan DFS dengan depth_limit=2. Artinya, pencarian dibatasi sampai kedalaman tertentu dari node awal.

```python
pos = nx.spring_layout(G, seed=42)
```

Kode ini menentukan posisi setiap node pada grafik menggunakan spring_layout. Nilai seed=42 digunakan agar posisi graph tetap sama setiap kali program dijalankan.

```python
plt.figure(figsize=(6, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=2000,
    font_size=12
)
```

Kode ini menampilkan visualisasi graph. Node ditampilkan dengan label, ukuran node diperbesar, dan ukuran font diatur agar mudah dibaca.
