## Penjelasan Dan Output Soal 2

```python
import matplotlib.pyplot as plt
import networkx as nx
```

Kode ini mengimpor library yang dibutuhkan. matplotlib.pyplot digunakan untuk visualisasi, sedangkan networkx digunakan untuk membuat dan menjalankan traversal pada graph.

```python
G = nx.DiGraph()

G.add_edges_from([
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("B", "E"),
    ("C", "E"),
    ("C", "F"),
    ("E", "F")
])
```

Kode ini membuat graph berarah menggunakan nx.DiGraph(). Edge yang ditambahkan memiliki arah, misalnya dari A ke B, dari C ke F, dan dari E ke F.

```python
start = "A"
goal = "F"

dfs_edges = list(nx.dfs_edges(G, source=start))
```

Kode ini menentukan node awal sebagai A dan node tujuan sebagai F. Setelah itu, DFS dijalankan dari node A dan hasil edge traversal disimpan dalam dfs_edges.

```python
print("Urutan DFS:")
for edge in dfs_edges:
    print(edge)
```

Kode ini mencetak urutan edge yang dikunjungi selama proses DFS. Output menunjukkan jalur penelusuran DFS dari node awal.

```python
visited_nodes = [start]

for edge in dfs_edges:
    visited_nodes.append(edge[1])
    if edge[1] == goal:
        break
```

Kode ini membuat daftar node yang sudah dikunjungi. Program menambahkan node dari hasil DFS satu per satu, lalu berhenti jika node tujuan F sudah ditemukan.

```python
print("\nUrutan traversal sampai goal ditemukan:")
for node in visited_nodes:
    print(node)
```

Kode ini menampilkan urutan node yang dikunjungi sampai goal ditemukan. Pada hasil program, traversal berhenti ketika node F berhasil dicapai.

```python
if goal in visited_nodes:
    print(f"\nGoal {goal} ditemukan menggunakan DFS")
else:
    print(f"\nGoal {goal} tidak ditemukan menggunakan DFS")
```

Kode ini mengecek apakah goal ada di dalam daftar node yang sudah dikunjungi. Jika ada, program menampilkan bahwa goal berhasil ditemukan menggunakan DFS.

```python
pos = nx.spring_layout(G, seed=42)
```

Kode ini menentukan posisi node dalam visualisasi graph menggunakan spring_layout. Nilai seed=42 membuat posisi node konsisten setiap kali program dijalankan.

```python
plt.figure(figsize=(6, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=2000,
    font_size=12,
    arrows=True
)
```

Kode ini menampilkan graph berarah dalam bentuk visual. Parameter arrows=True digunakan agar arah hubungan antar node terlihat pada grafik.
